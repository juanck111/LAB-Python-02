```python
# a. Múltiplo de 3
es_multiplo_de_3 = lambda x: x % 3 == 0

# b. Cubo
cubo = lambda x: x ** 3

# c. Producto
producto = lambda x, y: x * y

# d. Mayor de dos
mayor = lambda x, y: x if x > y else y

# e. Empieza con "A" o "a"
empieza_con_a = lambda palabra: palabra.lower().startswith('a')

# f. Celsius a Fahrenheit
celsius_a_fahrenheit = lambda c: (c * 9/5) + 32

# --- Aplicación de lista de lambdas ---
lambdas = [
    lambda x: x * 2,
    lambda x: x + 10,
    lambda x: x ** 2
]

numero = int(input("Ingrese un número: "))
for funcion in lambdas:
    print(funcion(numero))
```

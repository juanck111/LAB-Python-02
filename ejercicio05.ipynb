```python
lista_numeros = [2, 5, 8, 12, 15]

# a. Filtrar pares -> Elevar al cuadrado
pares_cuadrados = list(map(lambda x: x**2, filter(lambda x: x % 2 == 0, lista_numeros)))

# b. Elevar al cubo -> Filtrar > 100 -> Convertir a cadenas
cubos_str = list(map(str, filter(lambda x: x > 100, map(lambda x: x**3, lista_numeros))))

# c. Filtrar > 10 -> Elevar al cuadrado -> Ordenar mayor a menor
resultado_complejo = sorted(list(map(lambda x: x**2, filter(lambda x: x > 10, lista_numeros))), reverse=True)
```

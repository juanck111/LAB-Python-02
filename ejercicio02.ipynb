```python
# Datos de prueba para los ejercicios
numeros = [1, 2, 4, 11, 15]
palabras = ["Python", "java", "hola", "programacion", "sol"]
celsius = [0, 20, 100]
matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# a. Generar una lista con tuplas (n, n^2)
tuplas_cuadrados = [(n, n**2) for n in numeros]

# b. Cubos de los números mayores a 10
cubos_mayores_10 = [n**3 for n in numeros if n > 10]

# c. Solo los números pares
pares = [n for n in numeros if n % 2 == 0]

# d. ["valor: 10", "valor: 15", ...]
valores_str = [f"valor: {n}" for n in numeros]

# e. Conversión a grados Fahrenheit
fahrenheit = [(c * 9/5) + 32 for c in celsius]

# f. Filtrando palabras de más de 5 letras
palabras_largas = [p for p in palabras if len(p) > 5]

# g. Solo las iniciales de las palabras
iniciales = [p[0] for p in palabras]

# h. Indicando "par" o "impar"
paridad = ["par" if n % 2 == 0 else "impar" for n in numeros]

# i. Aplanar matriz
matriz_aplanada = [elemento for fila in matriz for elemento in fila]
```

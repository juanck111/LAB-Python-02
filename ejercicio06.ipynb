```python
import math

def minimax(piedras, es_max):
    # Condición de parada: Si no quedan piedras, el jugador que tiene el turno pierde.
    if piedras == 0:
        return -1 if es_max else 1

    if es_max:
        mejor_valor = -math.inf
        for movimiento in [1, 2, 3]:
            if piedras - movimiento >= 0:
                valor = minimax(piedras - movimiento, False)
                mejor_valor = max(mejor_valor, valor)
        return mejor_valor
    else:
        mejor_valor = math.inf
        for movimiento in [1, 2, 3]:
            if piedras - movimiento >= 0:
                valor = minimax(piedras - movimiento, True)
                mejor_valor = min(mejor_valor, valor)
        return mejor_valor

def mejor_movimiento_cpu(piedras):
    mejor_movimiento = 1
    mejor_valor = -math.inf
    for movimiento in [1, 2, 3]:
        if piedras - movimiento >= 0:
            valor = minimax(piedras - movimiento, False)
            if valor > mejor_valor:
                mejor_valor = valor
                mejor_movimiento = movimiento
    return mejor_movimiento

def jugar_nim():
    piedras = 10
    turno_min = True

    print("=== JUEGO NIM ===")
    print("Hay 10 piedras. Cada turno puedes quitar 1, 2 o 3.")
    print("Pierde el jugador que no tenga piedras para quitar.\n")

    while piedras > 0:
        print(f"Piedras restantes: {piedras} " + ("🪨 " * piedras))

        if turno_min:
            movimiento = 0
            while movimiento not in [1, 2, 3] or movimiento > piedras:
                try:
                    movimiento = int(input("Tu turno. ¿Cuántas quitas? (1-3): "))
                    if movimiento > piedras:
                        print("No puedes quitar más piedras de las que hay.")
                except ValueError:
                    print("Por favor, ingresa un número válido.")
            piedras -= movimiento
            turno_min = False
        else:
            print("Turno de la CPU...")
            movimiento = mejor_movimiento_cpu(piedras)
            print(f"La CPU decide quitar {movimiento} piedra(s).\n")
            piedras -= movimiento
            turno_min = True

    if turno_min:
        print("¡La CPU no tiene movimientos! ¡Has ganado!")
    else:
        print("¡No te quedan movimientos! La CPU gana.")


jugar_nim()
```
